<h3 style= "margin : 2"> Question 1 </h3>
Let PID be your student number, please use Python to calculate and explain each calculation:

    <li>b)PID//5;</li> 
    <li>c)PID%5;</li> 
    <li>d) PID ^2;</li> 
    <li>e) PID // 7;</li>

```python
PID = int(input("\n Enter your student number")) #Taking student number from user input and storing it in PID variable as an integer.

print(f"\n a) PID/5: {PID/5} -> In Python, the '/' operator is used for division and gives exact float value as a result")
print(f"\n b) PID//5: {PID//5} -> In Python, the '//' operator is used for division and gives only whole value as a result means value before decimal only")
print(f"\n c) PID%5: {PID%5} -> In Python, the '%' operator is used for getting the remainder after the division as a result")
print(f"\n d) PID^2: {PID**2} -> (PID^2) means PID to the power 2, to calculate power in Python we cannot calculate power using '^' directly instead we do it by using the '**' operator")
print(f"\n e) PID//7: {PID//7} -> n Python, the '//' operator is used for division and gives only whole value as a result means value before decimal only")
```

<h3 style= "margin : 2"> Question 2 </h3>
Let PID be a string generated by your student number, please use Python to do:
    <li>a) Print: first 3 letters of your PID;</li> 
    <li>b) Calculate the length of your PID;</li> 
    <li>c)Let aa be a string of the first 3 letters of your PID; and bb be a string of the last 3 letters of your PID; print a string with 6 letters with aa as the first 3 letters, and bb as the last 3 letters. (Hint: use aa + bb)</li> 
    <li>d) Append your PID to the string generated in question (c);</li> 
    <li>e) Calculate (23333*999/102)^3 using Python.</li>

```python
import math as m
PID = (input("\n Enter your student number")) #Taking student number from user input and storing it in PID variable as a string.
first_3_letters_PID = PID[:3] #finding the first 3 letters of PID and assigning them to a new variable
print(f"\n a) First 3 letters of PID is: {first_3_letters_PID}")

length_PID = len(PID) #calculating the length of PID and assigning it to a new variable
print(f"\n b) The length of the PID is: {length_PID}")

aa = PID[:3] #first 3 letters of your PID
bb = PID[-3:] #last 3 letters of your PID
aa_combined_bb = aa + bb #first 3 letters of your PID + last 3 letters of your PID
print(f"\n c) String with 6 letters with aa as the first 3 letters and bb as the last 3 letters is: {aa_combined_bb}")

aa_combined_bb_combined_PID = aa_combined_bb + PID #first 3 letters of your PID + last 3 letters of your PID + PID
print(f"\n d) String with PID to the string generated in question (c): {aa_combined_bb_combined_PID}")

print(f"\n e) Calculating (23333*999/102)^3: {(((23333*999)/102)**3)}") #Method 1

print(f"\n e) Calculating (23333*999/102)^3 using math library: {m.pow(((23333*999)/102),3)}") #Method 2
```

<h3 style= "margin : 2"> Question 3 </h3>
Using loops and conditional statements find the second largest number in the given user input list.

```python
numbers = list(map(int,input("Enter the list of numbers: ").split())) #Taking the list of numbers from the user input

largest = second_largest =float("-inf") #initially assigning largest and second_largest as close to -infinity

if(len(numbers)<2):
    print("For calculating the second largest number in list list should contain minimum 2 elements")
else:
    for number in numbers: #using 'for' loop to iterate through the list of numbers
        if number > largest: #using 'if' condition to check if value in number is greater than the value in largest
            second_largest = largest #updating value of second_largest with the value of largest
            largest = number #updating the value of largest with the value of number
        
        elif number > second_largest and number != largest: #using 'if' condition to check if value in number is greater than the value in second_largest and the value in number is not equal to the value in the largest
            second_largest = number #updating value of second_largest with value of number

    print(f"\n The second largest number in the given list of numbers is: {second_largest}")
```

<h3 style= "margin : 2"> Question 4 </h3>
Ask user to enter the mobile number
    <li>Use loops and conditional statements to add adjacent numbers and create a new string. Repeat this process until the number is reduced to a single-digit number.</li> 

```python
mobile_number = input("Enter your mobile number: ") #Taking the mobile number from the user input

def reduce_number(mobile_number):
    r_mobile_number = "" #creating an empty string
    for i in range (0, len(mobile_number) -1, 2): #interating through mobile number with step size = 2
        r_mobile_number += str(int(mobile_number[i]) + int(mobile_number[i+1])) #updating the empty string with value of 2 adjacent values of mobile numbers
        
    if len(mobile_number) % 2 != 0: #if size is odd add the last element as it is
            r_mobile_number+=mobile_number[-1]
        
    return r_mobile_number

if (len(mobile_number)!= 10): # mobile number should contain 10 digits
    print("Enter a valid mobile number containing 10 digits")

else:
    while(len(mobile_number)>1): #perform the operation till new length of mobile_number is greater than 1
        mobile_number = reduce_number(mobile_number)
    print(reduce_number(mobile_number)) 
```

<h3 style= "margin : 2"> Question 5 </h3>
You are given a list of student records, where each record is a dictionary containing the student's name, marks in various subjects, and attendance. The structure of each dictionary is as follows:
    <li>a) Calculate the average marks for each student across all subjects and store the result in a new key called "average_marks" within each student's dictionary.</li> 
    <li>b) Create a final grade for each student based on the following criteria: Store the grade in a new key called "grade" within each student's dictionary.</li> 
        <ul>
            <li>If the average marks are 90 or above and attendance is 90% or more, the grade is "A".</li>
            <li>If the average marks are between 80 and 89 and attendance is 85% or more, the grade is "B".</li>
            <li>If the average marks are between 70 and 79 and attendance is 80% or more, the grade is "C".</li>
            <li>Otherwise, the grade is "D".</li>
        </ul>
    <li>c) Output the names of students who have a grade of "A"</li>
   

```python
student_records = [
    {
        "name": "Abhinav",
        "marks": {"Physics": 95, "Chemistry": 96, "Maths": 92, "Biology":99},
        "attendance": 99
    },
    {
        "name": "Kareena",
        "marks": {"Physics": 99, "Chemistry": 86, "Maths": 82, "Biology":78},
        "attendance": 84
    },
    {
        "name": "Prabhjot",
        "marks": {"Physics": 79, "Chemistry": 63, "Maths": 60, "Biology":82},
        "attendance": 86
    },
    {
        "name": "Manik",
        "marks": {"Physics": 95, "Chemistry": 94, "Maths": 92, "Biology":93},
        "attendance": 97
    },
    {
        "name": "Sourabh",
        "marks": {"Physics": 68, "Chemistry": 65, "Maths": 73, "Biology":58},
        "attendance": 55
    },
    {
        "name": "Kishan",
        "marks": {"Physics": 54, "Chemistry": 38, "Maths": 48, "Biology":49},
        "attendance": 49
    },
    {
        "name": "Vivek",
        "marks": {"Physics": 89, "Chemistry": 56, "Maths": 92, "Biology":89},
        "attendance": 81
    },
    {
        "name": "Geetanjali",
        "marks": {"Physics": 89, "Chemistry": 85, "Maths": 84, "Biology":87},
        "attendance": 83
    },
    {
        "name": "Archi",
        "marks": {"Physics": 46, "Chemistry": 96, "Maths": 75, "Biology":85},
        "attendance": 57
    },
    {
        "name": "Richa",
        "marks": {"Physics": 89, "Chemistry": 67, "Maths": 43, "Biology":37},
        "attendance": 45
    }
]

#Function to calculate the average marks for each student
def calculate_average_marks(student_records):
    for student in student_records:
        marks = student["marks"] #storing each mark of the student in marks
        average_marks = sum(marks.values()) / len(marks) #calculating average marks by adding all the marks and then dividing it by the number of subjects
        student["average_marks"] = round(average_marks, 2) #rounding up the decimal value upto 2 decimal

calculate_average_marks(student_records) #calling the function to calculate average marks


#Function to calculate final grade for each student
def calculate_final_grade(student_records):
    for student in student_records:
        attendance = student["attendance"] #storing attendance of each student in attendance variable
        average_marks = student["average_marks"] #storing average marks of each student in average_marks variable
        if(average_marks>= 90 and attendance >=90):
            student["grade"]="A"
        elif(average_marks>= 80 and attendance >=85): 
            student["grade"]="B"
        elif(average_marks>= 70 and attendance >=80):
            student["grade"]="C"
        else:
            student["grade"]="D"

calculate_final_grade(student_records)  #calling the function to calculate final grade

#printing the name of students who got grade = "A"
print("\nStudents with Grade A are:")
for student in student_records:
    grade = student["grade"]
    if(grade == "A"):
        print(student["name"]) 
```
